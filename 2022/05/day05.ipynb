```python
from collections import deque
```

```python
from pathlib import Path
```

```python
t = Path('day05.txt').read_text().splitlines()
```

```python
t = """
    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2""".splitlines()
```

```python
# build a model of stacks and moves
stacks =  []
moves = []

for t2 in t:
    if "[" in t2:
        stacks.append(t2)
    elif t2.startswith("move"):
        moves.append(t2)
    elif t2.startswith(" 1 "):
        stackcount = int(t2.strip().split()[-1])
    else:    
        pass
```

```python
all_stacks = {k: deque() for k in range(1, stackcount +1)}
```

```python
for sl in stacks[::-1]:
    for st_nr, crate in enumerate(sl[1::4], start=1):
        if crate.strip():
            all_stacks[st_nr].append(crate)
        
```

```python
def serve_moves(move_list):
    for m in move_list:
        repeat, move_from, move_to = m.split()[1::2]
        for _ in range(int(repeat)):
            yield int(move_from), int(move_to)
        
```

```python
list(serve_moves(moves))
```

```python
def do_move(move_from, move_to):
    all_stacks[move_to].append(all_stacks[move_from].pop())
    
```

```python
for move_from, move_to in serve_moves(moves):
    do_move(move_from, move_to)
```

```python
def serve_moves2(move_list):
    for m in move_list:
        repeat, move_from, move_to = m.split()[1::2]
        yield int(move_from), int(move_to), int(repeat)
```

```python
def do_move2(move_from, move_to, move_count):
    tmp_stack = [all_stacks[move_from].pop() for _ in range(move_count)]
    assert len(tmp_stack) == move_count 
    print('stack', tmp_stack)
    for crate in tmp_stack[::-1]:
        all_stacks[move_to].append(crate)
```

```python
for move_from, move_to, move_count in serve_moves2(moves):
    print(move_from, "->", move_to, "x", move_count)

    do_move2(move_from, move_to, move_count)
    for i, line in all_stacks.items():
        print(i, list(line))
    print()    
```
