```python
t = """1000
2000
3000

4000

5000
6000

7000
8000
9000

10000""".splitlines()
```

```python
def calos(t):
    cals = 0
    elves = 0
    for c in t:
        # print(c)
        if not c:
            elves += 1
            yield elves, cals
            cals = 0
        else:
            cals += int(c)
    elves += 1            
    yield elves, cals            
```

```python
list(calos(t))
```

```python
x = list(calos(t))
```

```python
sum([t3[1] for t3 in sorted(x,  reverse=True, key=(lambda x: x[1]))[:3]])
```
